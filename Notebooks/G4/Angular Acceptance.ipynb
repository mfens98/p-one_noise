```python
import pandas as pd
import numpy as np
%matplotlib nbagg
import matplotlib.pyplot as plt
```

```python
angles = pd.read_csv("../acceptance_sim/build/acceptanceTest/analysis_QE.txt",delimiter="\t",header=None)
#SUMMER_STUDENS/mens/G4/acceptance_sim/acceptance_test/analysis_QE.txt
angle = angles[0]
hits = angles[1]
```

```python
plt.close()
plt.figure()
plt.plot(angle,hits/max(hits),'--o',label="Simulation")
plt.plot(angle,np.cos((0.5*angle)*np.pi/180)**4,label=r'$cos\left(0.5\cdot\theta\right)^4$')
plt.title("Simulated angular acceptance")
plt.ylabel("Normalized Intensity")
plt.xlabel("Angle (degrees)")
plt.legend(loc=0)
plt.show()
```
